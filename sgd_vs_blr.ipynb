```python

theta_values = np.linspace(-2, 10, 4000)
loss_values = modified_logistic_loss(theta_values)

# Adjusting the plot function to include multiple trajectories for different sigma values
# Let's say we have 3 different sets of m_values corresponding to 3 different sigma values

# Mock-up of three sets of m_values to represent different sigma values
sigmas = [1, 2, 3]  # Different sigma values
m_values_sets = [m_values1, m_values2, m_values3]

# Colors for different sigma trajectories
colors = ['red', 'green', 'blue']

# Begin plotting
plt.figure(figsize=(10, 6))

# Plot the loss function
plt.plot(theta_values, loss_values, label='Loss Function', zorder=1, color='black')

# Plot each trajectory
for sigma, m_values, color in zip(sigmas, m_values_sets, colors):
    iteration_indices = np.linspace(0, max(loss_values), len(m_values))
    plt.scatter(m_values, max(loss_values) - iteration_indices, color=color, alpha=0.6, label=f'Sigma = {sigma}', zorder=2)

# Additional plot formatting
plt.title('Trajectory of m values over Iterations for Different Sigmas')
plt.xlabel('Theta (m values)')
plt.ylabel('Loss Value')
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def combined_function(theta):
    # Minimum sharp via une fonction gaussienne centrée à -1 avec une petite variance
    sharp = np.exp(-30 * (theta + 1)**2)
    
    # Minimum flat via une fonction cosinus modifiée plus large et moins profonde
    flat = 0.3 * (1 + np.cos(np.pi * (theta -0.8) / 2.5 ))
    
    # Normalisation pour garder les valeurs entre 0 et 1
    combined = sharp + flat
    combined_norm = (combined - combined.min()) / (combined.max() - combined.min())
    
    return -combined_norm

# Générer des valeurs de theta
theta = np.linspace(-3, 3, 400)

# Calculer les valeurs de la fonction combinée
values = combined_function(theta)

# Afficher la fonction
plt.figure(figsize=(10, 5))
plt.plot(theta, values, label='Fonction avec Minima Sharp et Flat')
plt.ylim(-1, 0)
plt.legend()
plt.title('Fonction Unique avec Deux Minima Locaux')
plt.xlabel('Theta')
plt.ylabel('Valeur')
plt.show()
```
